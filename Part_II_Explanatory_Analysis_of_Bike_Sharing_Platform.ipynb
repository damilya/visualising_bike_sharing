## Investigation Overview


> The goal of this presentation is to show how various metrics of a bike sharing platform affect each other when we put them together. I will start with a simple distribution of the univariate explanatory distribution, then add up one and then more variables. It will show how new variables change the trends and help us to derive new insights.


## Dataset Overview

> The dataset consists of information regarding 183,000 records from a bike sharing platform Ford GoBike operating in San Fransisco Bay area. It describes the metrics on each ride such as duration in seconds, start and end time, stations names, member types, gender of the users, their age, and some other features. Around 800 records were removed due to possible data corruption related to the age of users and extremely long rides.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_bikes_data = pd.read_csv('fordgobike-tripdata.csv')
```

```python
# data wrangling to clean and prepare data
# removing null values
df_bikes_data.dropna(subset = ['start_station_id', 'start_station_name', 'end_station_id', 'end_station_name', 'member_birth_year', 'member_gender'], 
                    inplace = True)
# creating a new column with member age
df_bikes_data['member_age'] = 2019 - df_bikes_data['member_birth_year']
# removing geo locations 
df_bikes_data.drop(['start_station_latitude', 'start_station_longitude', 'end_station_latitude', 'end_station_longitude', 'member_birth_year'], axis = 1, inplace = True)
```

```python
# removing extreme outliers
df_bikes_data = df_bikes_data.query('duration_sec < 3000')
df_bikes_data = df_bikes_data.query('member_age < 70')
```

```python
# changing dtypes 
df_bikes_data['member_age'] = df_bikes_data['member_age'].astype(int)
df_bikes_data['user_type'] = df_bikes_data['user_type'].astype('category')
df_bikes_data['member_gender'] = df_bikes_data['member_gender'].astype('category')
```

```python
# helper function to title the graphs
def title_graph(main_title, x_title, y_title):
    plt.title(main_title)
    plt.xlabel(x_title)
    plt.ylabel(y_title)
```

## Distribution of Age 

> Member's age distribution has a range between 18 and 70 years old. The graph is right skewed, with a peak at 31 years old. The majority of the rides were done by the members between 25 and 35 years old. Some local peaks at 26 and 39 years can be spotted.

```python
plt.figure(figsize = [16, 5])
base_color = sb.color_palette()[0]
ax = sb.countplot(data = df_bikes_data, x = 'member_age', color = base_color)

title_graph('Distribution of members\' age\n', 
            '\nMember Age', 
            'Number of individual rides\n')

ax.set(yticklabels = (['0', '2k', '4k', '6k', '8k', '10k']))

plt.show()
```

## Age vs. Duration Per Ride

> Members' age has a direct impact on the duration per ride. Members in their end 20s and early 30s contribute to the longest rides the most compared to other age groups. Starting from about 33 years the duration of ride per age group descends with very few data points for 60 years olds and older. Interestingly, members in their mid and end 20s had a few long rides of 40-50 minutes, while no other age group rode longer than 43 minutes.

```python
plt.figure(figsize = [16, 5])
plt.hist2d(data = df_bikes_data, x = 'member_age', y = 'duration_sec', cmin = 60, cmap = 'viridis_r')

title_graph('Age vs. Duration Per Ride\n', 
            '\nMembers\' Age', 
            'Duration per ride (seconds)\n');

c_bar = plt.colorbar(label = '\nNumber of Rides')

plt.xticks([18, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000], ['0.5k', '1k', '1.5k', '2k', '2.5k', '3k'])

plt.show()
```

## Member Age to Member Gender and User Type

> The majority of rides gender-wise for both Customer and Subscription user types were done by 30-35 years olds with females being in the younger range. The youngest and the oldest members of both user types or all three genders are respectively 18 and 70 years old. A curious trend can be spotted among Male Subscribers in their late 20s. They tend to take less rides compared to in their mid 20s and mid 30s. Another point of interest is within Other gender group of members. Other Subscribers tend to cycle more in their mid 40s to 60, while Other Customers tend to be less active if they are in between 55 and 62, but tend to be more active if they are in between 62 and 70. The opposite is spotted at Other Subscribers, who cycle a lot in their late 40s to 60, and have almost no records for 60 plus age group.

```python
plt.figure(figsize = [16, 8]);
plt.switch_backend('agg');

g = sb.FacetGrid(data = df_bikes_data, col = 'member_gender', size = 5);
g.map(sb.violinplot, 'user_type', 'member_age', inner = 'quartile');

g.fig.subplots_adjust(top = 0.8);
g.fig.suptitle('Member Age to Member Gender and User Type');
g.set_titles(row_template = '{row_name}', col_template = '{col_name}');

g.set_xlabels('User Type');
g.set_ylabels('Member Age\n');

plt.show();
```
